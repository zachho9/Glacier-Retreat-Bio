```python
# from pygbif import species as species
# from pygbif import occurrences as occ

import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
df_sw = pd.read_csv("data/gbif_nz_snow_wren.csv", sep='\t')
df_sw.head()
```

```python
df_sw.info()
```

```python
df_sw = df_sw[["year", "individualCount", "decimalLatitude", "decimalLongitude"]]
df_sw.head()
```

```python
df_sw_yr = df_sw[df_sw['year'] == 2016.0]

gpd_sw_yr = (
    gpd.GeoDataFrame(
        df_sw_yr, geometry=gpd.points_from_xy(df_sw_yr.decimalLongitude, df_sw_yr.decimalLatitude), crs="EPSG:4326"
    )
)
```

```python
gpd_sw_yr.explore()
```

```python
df_sw_time_2 = df_sw[df_sw['year'] == 2022.0]

gpd_df_sw_time_2 = (
    gpd.GeoDataFrame(
        df_sw_time_2, geometry=gpd.points_from_xy(df_sw_time_2.decimalLongitude, df_sw_time_2.decimalLatitude), crs="EPSG:4326"
    )
)

# gpd_df_sw_time_2.explore()
```
